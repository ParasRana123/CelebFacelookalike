```python
import os
```

```python
actors = os.listdir('data')
```

```python
filenames = []
for actor in actors:
    for file in os.listdir(os.path.join('data' , actor)):
        filenames.append(os.path.join('data' , actor , file))
        
print(filenames)
print(len(filenames))
```

```python
import pickle
```

```python
pickle.dump(filenames , open('filenames.pkl' , 'wb'))
```
